```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.stats import binom
plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签
plt.rcParams['figure.figsize'] = (10.0, 8.0)  # set default size of plots
plt.rcParams['image.interpolation'] = 'nearest'
plt.rcParams['image.cmap'] = 'gray'


# 设置参数
p0 = 0.1  
# 设置真实概率范围
true_probs = np.arange(0.05, 0.16, 0.001)
for num,alpha in zip([609,370],[0.05,0.1]):
    # 计算功效
    powers = []
    critical_value = binom.ppf(alpha, num, p0) if alpha==0.05 else binom.ppf(1-alpha, num, p0) # 单侧检验的临界值

    for p in true_probs:
        # 计算功效为在备择假设下成功数小于临界值的概率
        power = binom.cdf(critical_value, num, p) if alpha==0.05 else binom.sf(critical_value, num, p)
        powers.append(np.round(power, 2))

    # result_df = pd.DataFrame({'真实概率': true_probs, '功效': powers})

    # 显示表格
    # print(result_df)

    # 绘制功效分析图
    plt.plot(true_probs, powers, linestyle='-', label=f"样本量={num}, 置信水平={alpha}")
plt.xlabel('真实概率')
plt.ylabel('功效')
plt.legend()
plt.title(f'两个原则下不同真实概率下的功效分析')
plt.grid(True)
plt.show()
```
